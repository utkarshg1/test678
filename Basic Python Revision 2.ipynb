# User defined function


```python
import math

def hypotenuse(a: float, b: float) -> float:
    """
    This function calculates hypotenuse for two sides of right
    angled triangle
    """
    c = math.sqrt(a**2 + b**2)
    return c
```

```python
h1 = hypotenuse(a=3, b=4)
print(h1)
```

```python
h2 = hypotenuse(a=10, b=11)
print(h2)
```

```python
help(hypotenuse)
```

# Generate square of first n natural numbers

```python
list(range(1, 10))
```

```python
list(range(1, 11))
```

```python
def gen_squares(n):
    for i in range(1, n+1):
        yield i**2
```

```python
gen_squares(10)
```

```python
list(gen_squares(10))
```

```python
for i in gen_squares(20):
    print(i)
```

```python
def gen_fib(n):
    # intilize first 2 terms
    a, b= 0, 1
    # Apply for loop 
    for _ in range(n):
        yield a
        a, b = b, a+b
```

```python
gen_fib(10)
```

```python
list(gen_fib(10))
```

```python
for i in gen_fib(20):
    print(i)
```

# Summarization and visualize

```python
# show average price of laptop by brand
a = (
    df.groupby(by = "brand")
    .agg({"Price": "mean"})
    .sort_values(by="Price")
    .round(2)
)
a
```

```python
year = df2["Order Date"].dt.to_period("Y")

yearly_sales = (
    df2.groupby(by = year)
    .agg({"Sales": "sum"})
    .round(2)
)

yearly_sales
```

```python
q = df2["Order Date"].dt.to_period("Q") 

quarterly_sales = (
    df2.groupby(by = q)
    .agg({"Sales": "sum"})
    .round(2)
)

quarterly_sales
```

```python
quarterly_sales.plot(kind="area", figsize=(8, 4))
```

```python
m = df2["Order Date"].dt.to_period("M")

monthly_sales = (
    df2.groupby(by = m)
    .agg({"Sales": "sum"})
    .round(2)
)

monthly_sales
```

```python
monthly_sales.plot(kind="line", figsize=(8, 4))
```

```python
w = df2["Order Date"].dt.to_period("W")

weekly_sales = (
    df2.groupby(by = w)
    .agg({"Sales": "sum"})
    .round(2)
)

weekly_sales
```
