```python
weight = float(input("Please enter weight in kgs : "))
height = float(input("Please enter height in meters : "))

bmi = weight / (height ** 2)
print(f"BMI : {bmi:.2f}")

if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi < 25:
    print("Normal Weight")
elif bmi >= 25 and bmi < 30:
    print("Overweight")
elif bmi >= 30 and bmi <= 35:
    print("Obesity 1")
elif bmi >= 35 and bmi < 40:
    print("Obesity 2")
else:
    print("Obesity 3")
```

# Show sum of elements inside a list

```python
nums = [12, 13, 11, 25, 101, 123, 145]

# Initilize sum with 0
s = 0

# Apply for loop on list
for i in nums:
    print(f"Current i : {i},Current Sum : {s}")
    s = s + i
    print(f"Updated sum : {s}")

# Print the final results
print(f"Sum of elements inside list {nums} is {s}")
```

# Function -> any input , output
p: princial, n: number of years, r: rate of interst

```python
def simple_interest(p: float, n: float, r: float) -> tuple:
    """
    p: Principal in INR
    N: Number of years
    R: Rate of Interst
    Output : Interest, Amount
    """
    i = (p*n*r)/100
    a = p + i
    return i, a
```

```python
i1, a1 = simple_interest(p=50000, n=5, r=10.5)
print(f"Interest : {i1}")
print(f"Amount : {a1}")
```

```python
i2, a2 = simple_interest(p=10_000, n=4, r=7.1)
print(f"Interest : {i2}")
print(f"Amount : {a2}")
```
